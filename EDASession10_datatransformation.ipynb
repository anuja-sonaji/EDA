$Data-Transformations $

- Generally data has 3 types of distribution
- Normal distribution
- Positive skewed
- Negative skewed

- All the maths concepts developed by assumptions as data follows ND
- Data skew happenes bcz of many reasons, in that one outliers
- Once we perform the outlier analysis then we can apply transformation techniques, if the data not follows normality

- - We have some transformation technologies
  - Log transformation
  - Exponential transformation
  - Reciprocal transformation
  - Power transformation
  - Squareroot transformation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
```

```python
exp_data = np.random.exponential(size=10000)
exp_data
```

```python
pt.hist(exp_data,bins=20,label="Exponential")
pt.show()
```

```python
norm_data = np.random.normal(size=10000)
norm_data
```

```python
pt.hist(norm_data,bins=20,label="Normal")
pt.show()
```

**Goal : converting exponential data [exp_data] to normal distribution**

**Log transformation**

  - Performing logarithm operations on original data

  - It is an approch to convert data into normal dist

  - Log means natural logarithm base=e

```python

logdata= np.log(exp_data)
```

```python
pt.figure(figsize=(10,4))
pt.subplot(1,2,1).hist(exp_data,bins=30,label="Exponential")
pt.legend()

pt.subplot(1,2,2).hist(logdata,bins=30,label="LogData")
pt.legend()
pt.show()
```

**Reciprocal Transformations**

```python
x=2
reci=1/x
reci
```

```python
reciprocal_data = np.reciprocal(exp_data)
reciprocal_data
```

```python
pt.figure(figsize=(10,4))
pt.subplot(1,2,1).hist(reciprocal_data,bins=30,label="Reciprocal")
pt.legend()
pt.show()
```

- When values are low reciprocal will fail
- for ex x=0 then 1/0 fail

```python
square_root_data = np.sqrt(exp_data)
square_root_data
```

```python
pt.figure(figsize=(10,4))
pt.subplot(1,2,1).hist(square_root_data,bins=30,label="Square-Root-Data")
pt.legend()
pt.show()
```

```python
pt.figure(figsize=(12,4))
pt.subplot(1,4,1).hist(exp_data,bins=30,label="Exponential")
pt.legend()

pt.subplot(1,4,2).hist(logdata,bins=30,label="LogData")
pt.legend()

pt.subplot(1,4,3).hist(reciprocal_data,bins=30,label="Reciprocal")
pt.legend()

pt.subplot(1,4,4).hist(square_root_data,bins=30,label="Square-Root-Data")
pt.legend()

pt.show()
```

```python
from sklearn.preprocessing import PowerTransformer

n= np.random.randn(100,5) * 5

pow_tr= PowerTransformer(method='yeo-johnson', standardize=True)

pow_tr.fit(n)

x_transform = pow_tr.transform(n)

x_original_scale=pow_tr.inverse_transform(x_transform)

x_original_scale

pt.figure(figsize=(12,4))
pt.subplot(1,2,1).hist(x_transform,bins=30,label="Transformed")
pt.legend()

pt.subplot(1,2,2).hist(x_original_scale,bins=30,label="Priginal")
pt.legend()
pt.show()
```
