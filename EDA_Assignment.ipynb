```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sns
```

```python
visadf=pd.read_csv('C:/Users/Anuja_PC/OneDrive/Documents/dataFiles/Visadataset.csv')
visadf
```

```python
visadf['education_of_employee']
```

```python
employeDf = visadf['education_of_employee'].value_counts()
employeDf
```

```python
empKeys=employeDf.keys()
empKeys
```

```python
empValues=employeDf.values
empValues
```

```python
employeeDf=pd.DataFrame(zip(empKeys,empValues),columns=['education_of_employee','count'])
employeeDf
```

$Bar Chart$

```python
#pt.figure(figsize=[10,5],facecolor='#EDB120')
pt.bar("education_of_employee","count",data=employeeDf)
pt.title("Employee Bar Chart")
pt.xlabel("--------Education of employee----------")
pt.ylabel("--------count---------")
pt.savefig("barchartemp.png")

pt.show()
```

$Seaborn Bar Chart$

```python
empKeys=employeDf.keys()
empKeys
```

```python
empKeys=employeDf.keys()
#pt.figure(figsize=[10,5],facecolor='#EDB120')
sns.countplot(data=visadf,x="education_of_employee", order=empKeys)
pt.title("Employee Seaborn Bar Chart")
pt.xlabel("--------Education of employee----------")
pt.ylabel("--------count---------")
pt.savefig("snsbarchartemp.png")

pt.show()
```

```python
ax=employeDf.plot(kind='bar')
ax.bar_label(ax.containers[0])
pt.show()
```

$Subplots$

```python
pt.figure(figsize=[10,5])
pt.subplot(1,2,1)
pt.bar("education_of_employee","count",data=employeeDf)
pt.title("Employee Bar Chart")
pt.xlabel("--------Education of employee----------")
pt.ylabel("--------count---------")
pt.savefig("barchartemp.png")


edf=visadf['education_of_employee'].value_counts()
keysInOrder=edf.keys()   # keys are in order
pt.figure(figsize=[10,5])
sns.countplot(data=visadf,x='education_of_employee', order=keysInOrder)
pt.title("sns Bar chart")
pt.xlabel("--------Continents----------")
pt.ylabel("--------count---------")
pt.savefig("snsbarchart.png")

```
