$Advantages$

- In Machine Leaening we have one important property is **independent of variables**.

- ML models expects variables are **independent of each other**

- One hot encoder creates new columns, which are **independent of each other**

- Independent each othervariables means both variables have **90 degree phase shift**

- 90 degree phase shift means **perpendicular to each other**

- Perpendicular means **orthogonal to each other** 

$Disadvantage$

- One hot encoder create new columns equal to no of unique lables

- if no of unique lables increases, new columns also increases

- More columns leads to more data

- id data is more complexity will increase

- complexity is more time increases for computation

- Entire process named as **Curse of dimensionality**
